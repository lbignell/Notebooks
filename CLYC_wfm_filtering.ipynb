# Waveform Filtering by using a Gaussian Process

So I've taken data using the CLYC detector and the pulses are *tiny*.
This means that the gamma spec energy resolution is bad (for CLYC) -- we'd expect ~5% at 662 keV whereas we get ~10%.
I think I can get some of this back with better waveform processing.
There are a couple of things to consider:

- The pulses are small (low S/N).
- There are random 'spikes' in the waveforms. I'm not sure about the cause of these, but they appear to be coincident on all channels of the DAQ. There is some mention in the DAQ driver code of functions to 'remove spikes', so presumably this is a known problem.
- There is the potential for a baseline offset.
- Ideally, whatever filtering/waveform processing I do shouldn't change the pulse shape significantly. I want to do PSD on these waveforms after all (optimising that is a warm up for the real thing in NaI).

So I think a Gaussian process interpolator may be the best option, off hand. This model essentially assumes the following:

- The waveform is a realisation of a Gaussian process that has a mean of zero and a covariance matrix that is specified.
- The covariance matrix functional form can be fit by choosing the maximum likelihood estimates of the parameters, given the waveform.
- The above two steps give a nice interpolation of the waveform.
- To make a noise-reducing filter, include a fixed white noise term in the waveform model. The noise can in principle be estimated by measuring the noise level when there is no signal, though in pracitice it may be easier to just tune it until the marginal likelihood is maximised.

Here's some waveforms to show what I mean. Note that the ordinate units are mV, which illustrates how tiny the pulses are...

```python
import tables
import json
from scipy import stats
import time as timemod

%pylab
%matplotlib inline

with open("RunIndex.json", 'r') as fh:
    RunIdx = json.loads(fh.readline())
    
sabrepath = '/data/SABRE/ANU-20170512/'
flist = sorted(RunIdx.keys())
h5file = tables.openFile(sabrepath+flist[40], 'r')
#Plot the first CLYC waveform for this file.
nwfms = 5
time = np.array([[0.]*1024]*nwfms)
wfm = np.array([[0.]*1024]*nwfms)
for i in range(nwfms):
    time[i] = h5file.root.time[i,0]
    wfm[i] = h5file.root.wfms[i,0]
    plt.plot(time[i], wfm[i])
    plt.ylim(-30, 10)
    plt.xlim(0, 400)

h5file.close()
```

### Rather than dive right in to using Gaussian process interpolation, it's probably prudent to look for simpler options. FFT comes to mind, and I'll look at others.

```python
#Let's have a look at the fft of these...
from scipy.fftpack import rfft, irfft, fftfreq
idx = 0
W = fftfreq(wfm[idx].size, d=time[idx,1]-time[idx,0])
f_wfm = rfft(wfm[idx])
#Filter the sucker
f_wfm_filt = f_wfm.copy()
f_wfm_filt[(W<0)|(W>0.2)] = 0
wfm_filt = irfft(f_wfm_filt)
plt.figure(figsize=(18,6))
plt.subplot(121)
plt.plot(time[idx], wfm[idx])
plt.plot(time[idx], wfm_filt, 'r')
plt.title('signal')
plt.xlabel('time (ns)')
plt.ylabel('voltage (mV)')
plt.xlim(0,400)
plt.subplot(122)
plt.plot(W, f_wfm)
plt.title('FFT of signal')
plt.xlabel('freq (GHz)')
plt.xlim(-0.1,0.35)
plt.ylim(-300,1300)
plt.grid()
```

```python
#Try a median filter:
from scipy.signal import medfilt
idx = 0
medfilt_wfm = medfilt(wfm[idx], kernel_size=7)
plt.plot(time[idx], wfm[idx], 'b')
plt.plot(time[idx], medfilt_wfm, 'r')
plt.xlim(0,300)
```

### Wow, that actually looks really good (and removes the spikes to boot).

The Gaussian process I mentioned above seems like it might be a special case of a Wiener filter (from the wikipedia article)...

```python
#Let's try a Wiener filter!
from scipy.signal import wiener
idx = 0
wiener_wfm = wiener(wfm[idx])
plt.plot(time[idx], wfm[idx], 'b')
plt.plot(time[idx], wiener_wfm, 'r')
plt.xlim(0,400)
#duh, the Wiener filter tries to preserve signal, and it's treating the spikes as signal!
#Lets wiener the median filtered-output:
wienmed_wfm = wiener(medfilt_wfm)
plt.plot(time[idx], wienmed_wfm, 'g')
plt.plot(time[idx], medfilt_wfm, 'm')
#OK, what about on the FFT?
wienFFT_wfm = wiener(wfm_filt)
plt.plot(time[idx], wienFFT_wfm, 'c')
```

```python
#Lets compare the wiener-median and just median in a bit more detail, shall we?
plt.plot(time[idx], wfm[idx], 'b')
plt.plot(time[idx], medfilt_wfm, 'r')
plt.plot(time[idx], wienmed_wfm, 'g')
plt.xlim(50,300)
plt.ylim(-18,-2)
plt.figure()
plt.plot(time[idx], wienmed_wfm-medfilt_wfm)
plt.xlim(50,300)
print(np.std(wienmed_wfm-medfilt_wfm))
```

```python
#Let's try this one:
fname = 'run1494029410.h5'
h5file = tables.openFile(sabrepath+fname)

def getQ(wfm,fast=None):
    '''
    Return the charge in the waveform.
    If 'fast' is given it must be an integer.
    '''
    BG = np.mean(wfm[10:30])
    if fast is None:
        Q = sum(wfm[40:]) - (1024-40)*BG
    else:
        Q = sum(wfm[40:fast]) - (fast-40)*BG
    return Q

#There's probably a more efficient way to do the filtering, but for now I'll do it one wfm at a time...
Nwfms = RunIdx[fname]['Nwfms']
Qmedian = np.array([0.]*Nwfms)
Qmedwien = np.array([0.]*Nwfms)
filt1 = np.array([0.]*1024)
filt2 = np.array([0.]*1024)
Qraw = np.array([0.]*Nwfms)
for i in range(Nwfms):
    filt1 = medfilt(h5file.root.wfms[i,0], kernel_size=7)
    filt2 = wiener(filt1)
    Qmedian[i] = getQ(filt1)
    Qmedwien[i] = getQ(filt2)
    Qraw[i] = getQ(h5file.root.wfms[i,0])

h5file.close()
```

```python
def loglikelihood_doublegauss(thedata, params):
    thevals = np.log((stats.exponweib.pdf(thedata, a=params['a'], c=params['c'], loc=params['mean'][0], scale=params['sigma'][0])*\
                          (params['amplitude'][0]) + \
                          stats.norm.pdf(thedata, loc=params['mean'][1], scale=params['sigma'][1])*\
                          (params['amplitude'][1])))
    return thevals if not np.iterable(thevals) else sum(thevals[thevals!=-np.inf])
```

```python
#def loglikelihood_doublegauss(thedata, params):
#    thevals = np.log((stats.norm.pdf(thedata, loc=params['mean'][0], scale=params['sigma'][0])*\
#                     (params['amplitude'][0]) + \
#                     stats.norm.pdf(thedata, loc=params['mean'][1], scale=params['sigma'][1])*\
#                     (params['amplitude'][1])))
#    return thevals if not np.iterable(thevals) else sum(thevals[thevals!=-np.inf])
def loglikelihood_doublegauss(thedata, params):
    thevals = np.log((stats.exponweib.pdf(thedata, a=params['a'], c=params['c'], 
                                          loc=params['mean'][0], scale=params['sigma'][0])*\
                          params['amplitude'][0]) + \
                          (stats.norm.pdf(thedata, loc=params['mean'][1], scale=params['sigma'][1])*\
                          (1-params['amplitude'][0])))
    return thevals if not np.iterable(thevals) else sum(thevals[(thevals!=-np.inf)&(thevals!=np.nan)])


#The data to be fit:
thedata = (Qfmedwien/Qtmedwien)[(-Qtmedwien<40000)&(25000<-Qtmedwien)&(Qfmedwien/Qtmedwien<0.475)&(Qfmedwien/Qtmedwien>0.275)]

#Initial guess of parameters:
params = {'mean': np.array([0.31, 0.435]), 
          'sigma': np.array([0.015, 0.01]), 
          'amplitude': np.array([4/7, 7]),
          'a': 1.6, 'c': 1.6}

ascend =True
ascendedlength = 50 #Number of trials that must fail to beat the likelihood before assuming we're at a maximum.

Nsteps = 10000
sigma_prop = 1e-6
ascending_sigma_scaling = 100
sigma_prop *= ascending_sigma_scaling

#Make space for output:
Trials = {'mean': np.empty([Nsteps, 2], dtype=np.float64),
          'sigma': np.empty([Nsteps, 2], dtype=np.float64),
          'amplitude': np.empty([Nsteps, 2], dtype=np.float64),
          'a': np.empty([Nsteps], dtype=np.float64),
          'c': np.empty([Nsteps], dtype=np.float64),
          'L_proposal': np.empty(Nsteps, dtype=np.float64),
          'L_current': np.empty(Nsteps, dtype=np.float64),
          'Ascending': np.empty(Nsteps, dtype=bool)}

starttime = timemod.time()

#This loop does some fit optimisation, then a MCMC starting at the optimal fit parameters.
for i in range(Nsteps):
    #store the trial
    for key, value in params.items():
        Trials[key][i] = value
    
    #Calculate the likelihood for the current guess:
    L_current = loglikelihood_doublegauss(thedata, params)

    if not i%1000:
        print("Step {0}, Likelihood = {1}".format(i, L_current))

    #Propose a new step and calculate the likelihood
    OK = False
    while not OK:
        newparams = {key: stats.norm.rvs(loc=value, scale=sigma_prop, size=2) for key, value in params.items() 
                     if key in ['mean', 'sigma', 'amplitude']}
        for key in ['a', 'c']:
            newparams[key] = stats.norm.rvs(loc=params[key], scale=sigma_prop)
        if newparams['amplitude'][0]>0.2:
            OK = True
    L_proposal = loglikelihood_doublegauss(thedata, newparams)
    
    #Do Monte Carlo to see if we take the step/do gradient ascent if ascending.
    if ascend:
        Trials['Ascending'][i] = True
        if L_proposal > L_current:
            NumConsecutive = 0
            params = newparams
        else:
            NumConsecutive += 1
            if NumConsecutive > ascendedlength:
                print("Starting MCMC after {0} trials".format(i))
                ascend = False
                sigma_prop /= ascending_sigma_scaling
                
    else:
        Trials['Ascending'] = False
        params = newparams if np.random.uniform() < np.exp(L_proposal - L_current) else params
        
    Trials['L_proposal'][i] = L_proposal
    Trials['L_current'][i] = L_current
    
endtime = timemod.time()
print('Finished after {0} seconds'.format(endtime-starttime))
```
