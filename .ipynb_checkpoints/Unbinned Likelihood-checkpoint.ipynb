# Unbinned Likelihood Fits and Homemade Markov Chain Monte Carlo

This notebook will make a simple unbinned likelihood fit, and implement a MCMC model.

```python
import numpy as np
import pymc
import matplotlib.pyplot as plt
import itertools
import os
import scipy as sp
import time
%matplotlib inline
```

```python
nsamples = 5000
#Let's make a 2D gaussian pdf as a sample:
mean = [0,1]
sigma = 2
corr = 0.3
cov = (sigma**2)*np.array([[1,corr],[corr,1]])
MV_dist = np.random.multivariate_normal(mean, cov, nsamples)
plt.subplot(1,2,1)
plt.plot(MV_dist[:,0], MV_dist[:,1], 'k.', markersize=1)
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.title("Unbinned Multivariate data")

#Let's also make a 1D PDF as a sample:
dist = sp.stats.exponnorm.rvs(K=1, loc=0, scale=1, size=nsamples)
plt.subplot(1,2,2)
plt.plot(dist, range(len(dist)), 'k.', markersize=1)
plt.xlabel("Value")
plt.title("Unbinned data")

plt.figure()
ret = plt.hist(dist, bins=100)
plt.title("1D data histogram")
plt.xlabel("Value")
```

Let's say we know that the data should be a MV gaussian.
The maximum likelihood method lets us estimate the parameters for unbinned data.
The likelihood of some parameter set $\Theta$ is given by:

$L\left(\Theta\right) = \prod_{i=1}^{n} f\left(x_{i} | \Theta\right)$

The negative log likelihood is computationally better to maximise (it avoids numbers that are too small/large):

$\log\left(L\left(\Theta\right)\right) = - \sum_{i=1}^{n} \log\left(f\left(x_{i} | \Theta\right)\right)$

It can be shown by smarter people than me that (presumably if you use the correct normalisation on the PDF) the $N\sigma$ uncertainty in a parameter is given by the width of the likelihood function at $\frac{N^{2}}{2}$ units below its peak.

Let's test this out on the single dimensional case (assume we know every parameter except the mean for now, for simplicity):

```python
#Scipy has a builtin normalised Gaussian PDF
sp.stats.exponnorm.pdf(0,K=1,loc=0,scale=1)#*np.sqrt(2*np.pi)
```

```python
#Lets scan over possible mean values
meanvals = np.linspace(-2, 2, 5001)
likelihoods = np.sum(np.log(sp.stats.exponnorm.pdf(dist, K=1, loc=meanvals[:,None], scale=1)), axis=1)
```

```python
maxidx = np.argmax(likelihoods)
onesigmaidx = np.argmax(likelihoods[likelihoods<(likelihoods[maxidx]-0.5)])
print(meanvals[onesigmaidx])
```

```python
#Let's plot this:
plt.plot(meanvals, likelihoods)
#find the minimum likelihood
plt.axhline(y=likelihoods[maxidx], color='k', linestyle='dashed')
plt.axvline(x=meanvals[maxidx], color='k', linestyle='dashed')
plt.axhline(y=likelihoods[onesigmaidx], color='r', linestyle='dotted')
plt.axvline(x=meanvals[onesigmaidx], color='r', linestyle='dotted')
plt.axvline(x=meanvals[maxidx -(onesigmaidx - maxidx)], color='r', linestyle='dotted')
plt.xlabel('Distribution Mean')
plt.ylabel('log-likelihood')
plt.title('mean = {0}, sigma (assuming symmetric) = {1}'.format(meanvals[maxidx], 
                                                                np.abs(meanvals[onesigmaidx] - meanvals[maxidx])))
print("Distribution sigma (expected) = {0}".format(sp.stats.exponnorm.std(K=1,loc=0,scale=1)))
print("Distribution sigma (inferred) = {0}".format(np.abs(meanvals[np.argmax(likelihoods[likelihoods<
                                                                         (likelihoods[maxidx]-0.5*len(dist))])])))
print("Distribution sigma (measured) = {0}".format(np.std(dist)))
plt.ylim((max(likelihoods)-5, max(likelihoods)))
plt.xlim(-0.5,0.5)
#plt.title('mean = {0}, sigma (assuming symmetric) = {1}\nE(mean) = {2}, E(sigma) = {3}'.format(meanvals[minidx], 
#                                                                np.abs(meanvals[onesigmaidx] - meanvals[minidx]),
#                                                                sp.stats.exponnorm.mean(K=1,loc=0,scale=1),
#                                                                sp.stats.exponnorm.std(K=1, loc=0, scale=1)))
```

OK, so I also figured out (quite by accident) that I can get an estimate of the distribution standard deviation can be estimated using the width of the likelihood function at $\frac{N_{samples}}{2}$ below the maximum likelihood. Bonus!

Consider the multivariate data. I think there are too many parameters to scan over to make a grid search practical: there's 2 means and 3 non-redundant elements in the covariance matrix. It's easy to imagine even more complicated models. Let's try a Markov Chain Monte Carlo (MCMC)!

MCMC is just like a random walk which does the following:

1. Initialise by randomly choose a parameter vector.

2. Calculate the probability of the current parameter vector state, $P_{current}$.

3. Propose to a step in the parameter vector space. How you take a step depends upon the algorithm you're using. The Metropolis algorithm just samples from a normal distribution centred on the current parameter vector with a width $\sigma_{proposal}$.

4. Calculate the probability of the proposed step, $P_{proposal}$.

5. The probability of taking a the proposed step is taken to be the ratio of the proposed to current likelihoods (this is the Monte Carlo part), i.e. $p_{step} = \frac{P_{proposal}}{P_{current}}$. Take the step if a random draw < $p_{step}$. 
    
    *Aside: This is genius! If you're far from the maximum, you'll always accept steps that improve the maximum likelihood. If you're near the maximum, you essentially do a random walk which is weighted by the posterior amplitude!* The reason for this (as explained [here](http://twiecki.github.io/blog/2015/11/10/mcmc-sampling/)) is that it's just Bayes' theorem with the normalisation part ($\int P(x,\Theta) dx$) ratioed out so it's just $P\left(x | \Theta_{proposal}\right)P\left(\Theta_{proposal}\right) / P\left(x | \Theta_{current}\right)P\left(\Theta_{current}\right)$. Note that below I'm implicitly assuming the prior $P\left(\Theta\right)$ is uniform.

6. If the maximum number of steps isn't exceeded, go to 2.

Let's make a MCMC!

```python
#Mean is a vector, var is assumed to be the same in all dims (so is a const), similarly corr is a constant.
true_params = {'mean': mean, 'sigma': sigma, 'corr': corr}
#1. Make an initial guess at the parameter vector:
params = {'mean': [-0.5,0.5], 'sigma': 1, 'corr': 0.0}

#Simulation parameters
Nsteps = 10000
sigma_proposal = 1e-3
step_enhance_factor = 1.0
peakfind_sigma_scaling = 100
#Start with bigger steps to reach to peak more quickly
sigma_proposal *= peakfind_sigma_scaling

#make space for output
Trials = {'mean': np.empty([Nsteps, len(mean)], dtype=np.float64), 
          'sigma': np.empty(Nsteps, dtype=np.float64),
          'corr': np.empty(Nsteps, dtype=np.float64),
          'L_proposal': np.empty(Nsteps, dtype=np.float64),
          'L_current': np.empty(Nsteps, dtype=np.float64),
          'Ascending': np.empty(Nsteps, dtype=bool)}

ascend = True
ascendedlength = 50 # Once 100 trials can't beat the likelihood, let's assume we've found the right params.
NumConsecutive = 0

starttime = time.time()

for i in range(Nsteps):
    
    if not i%1000:
        print("Up to step {0}".format(i))
        print("Params: {0}".format(params))
    #Store this trial
    for key, value in params.items():
        Trials[key][i] = value
    
    currentcov = (params['sigma']**2)*np.array([[1,params['corr']],[params['corr'],1]])
    
    #2. Calculate the likelihood of the current guess
    L_current = np.sum(sp.stats.multivariate_normal.logpdf(MV_dist, mean=params['mean'], cov=currentcov))
    
    if not i%1000:
        print("L_current: {0}".format(L_current))

    #3. Propose the new step:
    ok = False
    while not ok:
        newparams = {key: sp.stats.multivariate_normal.rvs(mean=value, cov=sigma_proposal) 
                         for key, value in params.items()}
        if newparams['corr']>-1 and newparams['corr']<1:
            ok = True
    
    if not i%1000:
        print("newparams: {0}".format(newparams))

    newcov = (newparams['sigma']**2)*np.array([[1,newparams['corr']],[newparams['corr'],1]])
    
    #4. Calculate the likelihood for the new step
    L_proposal = np.sum(sp.stats.multivariate_normal.logpdf(MV_dist, mean=newparams['mean'], cov=newcov))
    if not i%1000:
        print("L_proposal: {0}".format(L_proposal))
    
    #5. Do a Monte Carlo trial to see if we take the step.
    #EDIT: the convergence takes forever, so let's do what is effectively a gradient ascent first; then transition to MCMC.    
    if ascend:
        Trials['Ascending'][i] = True
        if L_proposal > L_current:
            NumConsecutive = 0
            params = newparams
        else:
            NumConsecutive += 1
            if NumConsecutive > ascendedlength:
                print("Starting MCMC after {0} trials.".format(i))
                ascend = False
                sigma_proposal /= peakfind_sigma_scaling
                
    else:
        Trials['Ascending'][i] = False
        #params = newparams if np.random.uniform()*step_enhance_factor < L_current/L_proposal else params
        params = newparams if np.random.uniform()*step_enhance_factor < np.exp(L_proposal-L_current) else params
    
    Trials['L_proposal'][i] = L_proposal
    Trials['L_current'][i] = L_current
    
endtime = time.time()
print("Finished after {0} seconds".format(endtime-starttime))
```

```python
offset = 0#sum(Trials['Ascending'])
plt.plot(range(Nsteps - offset), Trials['mean'][offset:])#, label=['mean_0', 'mean_1'])
plt.plot(range(Nsteps - offset), Trials['corr'][offset:])
plt.plot(range(Nsteps - offset), Trials['sigma'][offset:])
plt.figure()
plt.plot(range(Nsteps - offset), Trials['L_proposal'][offset:])
plt.plot(range(Nsteps - offset), Trials['L_current'][offset:])

offset = sum(Trials['Ascending'])
plt.figure()
key = 'sigma'
plt.hist(Trials[key][offset:], bins=100)
plt.title("Histogram of {0}".format(key))

plt.figure()
plt.plot(Trials['mean'][offset:,0], Trials['mean'][offset:,1], 'k.', markersize=2)
plt.xlabel('mean_0')
plt.ylabel('mean_1')
```

Wow, that was more complicated than I was expecting!
There were two key things I realised:

- You must compare the *likelihood* ratio, not the *log-likelihood* ratio. For a MCMC, the parameters in the ratio must be proportional to the likelihood $L\left(\Theta | x\right)$. Note that this can be difficult because the numbers can be small. The trick is to use log-likelihood until right at the end and use 64 bit floats, not 16 bit (which I was doing to try and be 'clever' and speed things up...)

- A poor initial parameter guess can screw up a MCMC. The numbers can just be too small to compute the probabilities. Plus convergence can be slow. I got around this by including an initial phase where I just did essentially a random walk with gradient ascending steps only allowed. This initial phase was considered over when N consecutive steps didn't improve the log-likelihood ratio. This means the MCMC converges essentially straight away and (hopefully) doesn't run into any big/small number issues.

- Choosing the parameter that governs the width of the proposal distribution is not entirely trivial. I don't see an easy way around this other than trial-and-error.

OK, so I've proven to myself that I can make a MCMC to fit unbinned data, which is pretty cool.
