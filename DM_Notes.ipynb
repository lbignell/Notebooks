## Freese *et al*., "Annual Modulation of Dark Matter: A Review" [ArXiV 1209.3339](https://arxiv.org/abs/1209.3339) (2013)

This paper is a review of direct DM modulation searches. The state of experiments at the time of writing was that there was some tension with the LXe experiments, but perhaps not as much as today.

#### *Why WIMPs are favoured DM candidate*:
If there are WIMPs in the 1 GeV -- 10 TeV mass range; their fractional contribution to the energy density of the universe ($\Omega_{\chi}$) is:

$\Omega_{\chi}h^{2} \sim \left(3 \times 10^{-26} \text{cm}^{3}/\text{sec} \right)/\langle \sigma v \rangle_{ann}$

where $\langle \sigma v \rangle_{ann}$ is the DM annihilation cross-section. A value for $\langle \sigma v \rangle_{ann}$ consistent with the weak force automagically gives the right answer, near the values measured by astronomers (WMAP). This coincidence is known as the "WIMP miracle" and seems to be why WIMPs are the preferres DM candidate.

### Recoil Rate

- Assumes WIMP elastic scattering as the detection paradigm. Note that (as the DAMA folks point out) elastic scattering is not the only signal which can be measured by modulation experiments.

- The differential nuclear recoil rate with a nucleus with mass M is: 

 $\frac{dR}{dE_{nr}} = \frac{n_{\chi}}{M}\langle v \frac{d\sigma}{dE_{nr}} \rangle = \frac{2\rho_{\chi}}{m_{\chi}}\iiint v f\left(\vec{v},t\right) \frac{d\sigma}{dq^{2}}\left(q^{2}, v\right) d^{3}v$
 
 where $n_{\chi} = \rho_{\chi}/m_{\chi}$ is the DM number density ($\rho_{\chi}$ is the DM mass density), $f\left(\vec{v},t\right)$ is the time-dependent DM velocity distribution (relative to the recoiling nucleus), and $\frac{d\sigma}{dq^{2}}\left(q^{2}, v\right)$ is the velocity-dependent differential cross-section, with $q^{2} = 2ME_{nr}$ the lab-frame momentum exchange in the scatter. The conventional coupling assumptions give rise to an expression:
 
 $\frac{dR}{dE_{nr}} = \frac{\rho_{\chi}}{2m_{\chi}\mu^{2}}\sigma\left(q\right)\eta\left(v_{min}\left(E_{nr}\right),t\right)$.
 
 where $\sigma\left(q\right)$ is an effective scatter cross-section and $\eta\left(v_{min}\left(E_{nr}\right),t\right)$ is the mean inverse speed. Written in this form, the particle physics and astrophysics contributions separate out into the cross-section and $\rho_{\chi}\eta\left(v_{min}\left(E_{nr}\right),t\right)$:
 
 $\eta\left(v_{min}\left(E_{nr}\right),t\right) = \iiint_{v>v_{min}} \frac{f\left(\vec{v},t\right)}{v} d^{3}v$
 
 is the mean inverse speed and:
 
 $v_{min} = \begin{cases}
       \sqrt{\frac{ME_{nr}}{2\mu^2}} &\text{(elastic)}\\
       \frac{1}{\sqrt{2ME_{nr}}}\left(\frac{ME_{nr}}{\mu} + \sigma\right) &\text{(inelatic)}\\
     \end{cases}$
 
 where $\mu = m_{\chi}M/\left(m_{\chi} + M\right)$ and $\delta$ is the mass splitting to the excited state.
 
- The paper then looks in more detail about the particle physics (cross-section) and astrophysics (DM density and velocity distribution) components.
 
#### WIMP elastic scatter cross-section

- For many WIMP candidates, the dominant couplings are the scalar and axial-vector couplings, which leads to a spin-independent (SI) and a spin-dependent (SD) cross-section.

- For both SI and SD; the velocity-dependent differential cross-section can be written as:

 $\frac{d\sigma}{dq^{2}}\left(q^{2}, v\right) = \frac{\sigma_{0}}{4\mu^{2}v^{2}}F^{2}\left(q\right)\Theta\left(q_{max} - q\right)$
 
 where $\Theta$ is the Heaviside step function, $\sigma_{0}$ is cross-section in $v \rightarrow 0$, $q_{max} = 2\mu v$ is the maximum momentum transfer, and $F^{2}\left(q\right)$ is a form factor to account for the finite size of the nucleus.
 
- At low $q$, the De-Broglie wavelength of the momentum transfer $>>$ the size of the nucleus, so $F^{2}\left(q\right) = 1$. However, at higher $q$ this approximation is no longer valid so the cross-section must be corrected for this form factor.

- For the SI cross-section, if the coupling to neutrons and protons is the same;

 $\sigma_{SI} = \frac{\mu}{\mu_{p}}A^{2}\sigma_{p,SI}$
 
 where $\mu_{p}$ is the 'proton-reduced mass' (it isn't clear to me what that is, but I think its a correction for treating neutrons the same as protons, even though they have different masses?), $A$ is the number of nucleons in the nucleus, and $\sigma_{p,SI}$ is the SI cross-section for a proton.
 
- Note that $\sigma_{SI}$ has a strong dependence on $A$, not just through the $A^{2}$ term: if $m_{\chi} >> M$, then $\frac{\mu}{\mu_{p}} \rightarrow A^{2}$. The form factor correction tempers this dependence somewhat, but all other things being equal, *heavier nuclei should have bigger SI cross-sections*.

- For reasons that aren't clear to me; apparently the form factor is just a Fourier transform of the mass distribution of the nucleus, and can be guesstimated as:

 $F\left(q\right) = 3e^{-q^{2}s^{2}/2}\frac{\sin\left(qr_{n}\right) - qr_{n}\cos\left(qr_{n}\right)}{\left(qr_{n}\right)^{3}}$
 
 where $s \sim 0.9$ fm, $r_{n}^{2} = c^{2} + \frac{7\pi}{3}a^{2} - 5s^{2}$ is the nuclear radius, with $a \sim 0.52$ fm and $c \sim 1.23 A^{1/3} - 0.6$ fm.
 
- The SD cross-section should only be non-zero for detector isotopes with an unpaired proton or paired neutron. Apparently it can be written as:

 $\sigma_{SD} = \frac{32\mu^{2}}{\pi}G_{F}^{2}J\left(J+1\right)\Lambda^{2}$
 
 where $G_{F}$ is the Fermi constant, J is the nuclear spin, and:
 
 $\Lambda = \frac{1}{J}\left(a_{p}\langle S_{p} \rangle + a_{n}\langle S_{n} \rangle\right)$
 
 where $\langle S_{p} \rangle$ ($\langle S_{n} \rangle$) and $a_{p}$ ($a_{n}$) are the effective spin contributions from the proton (neutron) groups and the SD couplings to the proton (neutron), respectively.
 
- Unlike in SI case, the SD couplings to protons and neutrons aren't usually thought to be identical.

- The spin factors aren't expected to scale with $A$, so the nucleon number dependence only goes with $\sim A^{2}$, not $\sim A^{4}$ like in SI case.

- Some theories predict much stronger SD couplings ($\sim 10^{2} - 10^{4}$ times) than SI couplings.

- SD form factors vary quite a bit from nucleus to nucleus.

- More general forms to the cross-section exist, which may introduce momentum-dependence. Because the momentum is usually thought to be small ($v \sim 10^{-3} c$), these are usually ignored, except in theories where the scalar and axial-vector couplings are suppressed. This can lead to a recoil spectrum that differs from what you'd expect from elastic scattering.

### Dark Matter Distribution:

- The DM rest frame is usually taken to be the same as the Galactic rest frame.
- Model estimates of $\rho_{\chi}$ differ by a factor of two or so, but $\rho_{\chi} \sim 0.4$ GeV/cm$^{3}$.
- The DM velocity distribution in the rest frame, $\widetilde{f}\left(\vec{v}\right)$, is taken to be an isotropic, isothermal, Maxwellian in the Standard Halo Model (SHM):

 $\widetilde{f}\left(\vec{v}\right) = \begin{cases}
       \frac{1}{N_{esc}}\left(\frac{3}{2\pi\sigma_{v}^{2}}\right)^{3/2}e^{-3\vec{v}^2/2\sigma_{v}^{2}} &\text{for} |\vec{v}| < v_{esc}\\
       0 &\text{otherwise}\\
    \end{cases}$

 where
 
 $v_{esc} = \text{erf}\left(z\right) - \frac{2}{\sqrt{\pi}}ze^{-z^{2}}$
 
 with $z = v_{esc}/v_{0}$ and $v_{0} = \sqrt{2/3}\sigma_{v}$ (the most probable speed, about 235 km/sec).
 
- In general, it seems hard to model the tail of the velocity distribution (because of leakage from DM escape from the galaxy).

- Combining the SHM with the usual form of SI and SD cross-sections, gives

 $\frac{dR}{dE_{nr}} \sim e^{-E_{nr}/E_{0}}$
 
 where $E_{0}$ is some scale factor usually taken to be $\sim 10$ keV.
 
- So the recoil spectrum drops off exponentially with energy.

- But... The SHM is only a first approximation.

- Ellipsoidal DM distributions lead to anisotropic velocity profiles, which can affect the modulation signal by $\sim 10\%$.

- Since the tail of the distribution is so uncertain, this means low-mass WIMP detection is more astrophysics model-dependent.

- The DM distribution depends on the history of the Milky Way's formation (mergers, DM sub-halo mergers). The sun may be gravitationally attracting a local over-density of DM, though apparently it can't be more than 15% of the average Milky Way density (according to the Acquarius Project). The history may lead to 'debris flows' and 'tidal streams' from past merger events that can affect local velocity distributions. There can also be extra-galactic DM...

## Annual Modulation

- The DM velocity distribution in the earth's frame is:

 $f\left(\vec{v}, t\right) = \widetilde{f}\left(\vec{v}_{obs}\left(t\right) + \vec{v}\right)$
 
 where
 
 $\vec{v}_{obs}\left(t\right) = \vec{v}_{\odot} + \vec{v}_{\oplus}\left(t\right)$
 
 with $\vec{v}_{\odot}$ the sun's velocity relative to the galactic rest frame. The main component of this velocity is the rotational velocity of our part of the disk, which is $\sim 235$ km/sec. 
 
 The Earth's velocity is just motion about a circle (neglecting orbital eccentricity):
 
 $\vec{v}_{\oplus} = v_{\oplus}\left[\hat{\epsilon}_{1}\cos\omega\left(t - t_{1}\right) + \hat{\epsilon}_{2}\sin\omega\left(t - t_{1}\right)\right]$
 
- But relative velocity just simplifies to (assuming SHM):

 $v_{obs}\left(t\right) \sim v_{\odot} + bv_{\oplus}\cos\omega\left(t - t_{1}\right)$
 
 where $b = \sqrt{b_{1}^{2} + b_{2}^{2}}$ and $b_{i} = \hat{\epsilon}_{i} \cdotp \vec{v}_{\odot}$.
 
- Which gives a recoil signal of:

 $\frac{dR}{dE_{nr}}\left(E_{nr}, t\right) \sim S_{0}\left(E_{nr}\right) + S_{m}\left(E_{nr}\right)\cos\omega\left(t - t_{0}\right)$
 
 where $S_{0}$ and $S_{m}$ are the time-averaged rate and the modulation amplitude, respectively.
 
- Because the rotational velocity of the earth (0.5 km/s) is so much smaller than its orbital velocity (30 km/s), the daily (diurnal) modulation is usually ignored. However, directional detectors may be able to see a daily modulation in the direction of the recoils.

- The SHM is only a first approximation, and having debris flows or streams can modify the recoil rate and the modulation amplitude in various ways. The paper goes into some detail about this, but I think that's enough detail for my purposes.
